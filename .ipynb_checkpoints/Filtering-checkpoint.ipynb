## condition filtering

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('tips.csv')
```

```python
df
```

```python
#  Single condition
bool_series=df['total_bill']>40
```

```python
df[bool_series]
```

```python
df[df['sex']=="Male"]
```

```python
#  Multiple conditions
# in python , and -> to compare two bool values
# in pandas , & -> to compare two series of bool values
df[(df['sex']=='Male') & (df['total_bill']>30)]
```

```python
df[(df['day']=='Sun') | (df['day']=='Sat') | (df['day']=='Fri')]
```

```python
#  Shortcut
options = ['Sat','Sun']
```

```python
df['day'].isin(options)
```

```python
df[df['day'].isin(options)]
```

```python
df[df['day'].isin(['Sat','Sun'])]
```
