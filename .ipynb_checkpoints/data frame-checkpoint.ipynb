```python
#  operations same as numpy array
df['total_bill']+df['tip']
```

```python
100*df['tip']/df['total_bill']
```

```python
# adding new column
df['tip_percentage']=100*df['tip']/df['total_bill']
```

```python
df
```

```python
df['price_per_person']=df['total_bill']/df['size']
# if this column is already present , then it will overwrite that column
```

```python
df
```

```python
# rounding off the value
df['price_per_person']=np.round(df['price_per_person'],2)
```

```python
df
```

```python
# dropping the column
df.drop('tip_percentage',axis=1)    # axis=0 -> row  axis=1 -> column
```

```python
df  #Feature of pandas -> these functions we implement are not in place (i.e. in original dataframe)
```

```python
# to change in place
df=df.drop('tip_percentage',axis=1)
```

## accessing based on numerical index & labelled index

```python
df=df.set_index('total_bill')
```

```python
df.iloc[0]  # numeric index access , based location in sequence
```

```python
df.loc[16.99]  # accessing from labelled index
```

```python
#  accessing multiple rows
df.iloc[0:4]
```

```python
df.loc[[16.99,10.34,21.01,23.68]]
```

```python
#  dropping particular row
df.drop(10.34,axis=0)
```

```python
df
```

```python
#  appending row
one_row=df.iloc[0]
```

```python
type(one_row)
```

```python
df._append(one_row, ignore_index=True)
```
